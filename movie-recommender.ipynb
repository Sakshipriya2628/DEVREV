## Popularity based recommendation

```python
#Filtering needed columns to find Popular Movies
df = df2[['id','title','vote_count', 'vote_average','popularity']]
```

```python
df
```

```python
#Giving a particular score  for each movies using three major numerical colums values *wiz: vote_count, vote_average, popularity
#For that a eqn derive to find the the Score for the movie
def ratings (x):
    vote = x['vote_count']
    average = x['vote_average']+1
    popularity = x['popularity']+1
    
    
    rating = (vote*popularity)/average
    log = np.log2(rating)
    return log
    
```

```python
df['score'] = df.apply(ratings, axis=1)
```

```python
#Recommending most Popular movies based on the data set
popular_movie = df[df.vote_count>=2000].sort_values('score',ascending=False).head(50)
```

```python
#Float values decimal numbers rounding off to 1 digit after decimal
popular_movie.round(1)
```

## Description(Content) based Recommender
 To find most 10 similar movies

```python
#Based on the taglines, genres, description we can concat them and convert it to a feature of vector for each  particular movie

df3 = df2[['id','title', 'genres', 'cast', 'crew','production_companies','keywords','overview']]
df3
```

```python
df3.genres.iloc[0] #and its a string not a list
```

```python
import ast
#By using this "ast" modules own  literal_eval function we can convert the  string to list
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Now I've the list of generes detail . Need to fetch all the genres of that movie (which is a good parameter to find similar movie)
#Converting the dictionary name values and appending towards a list

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
df3.genres = df3.genres.apply(convert)
```

```python
df3
```

```python
#To find 3 Major Lead Actor's name of  each movie

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
df3.cast = df3.cast.apply(convert3)
```

```python
df3.crew.iloc[0]
```

```python
#We can see that in crew column the full crew members associated with the movies are included in a single cell also these are messed with lots of unwanted details


#To extract the director name from each row we can use this function
#This function will only detect the job name=Director and append it to the list
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df3['director'] = df3.crew.apply(fetch_director)
```

```python
df3
```

```python
df3.productions = df3.production_companies.apply(convert)
```

```python
df3 = df3.drop(columns = ['crew','production_companies'])  # already created  with new name, so dropping the older columns
```

```python
df3.keywords =df3.keywords.apply(convert)
```

```python
df3
```

```python
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return [str.lower(x.replace(" ", ""))for i in x]
        else:
            return ''
```

```python
df3.info()
```

```python
# Apply clean_data function to your features.
features = ['cast', 'keywords', 'genres', 'director']

for feature in features:
    df3[feature] = df3[feature].apply(clean_data)
```

```python
#df3['overview'] = df3['overview'].fillna('')
#df3['overview']  = df3['overview'].astype('str')
#df3['overview']  = df3['overview'].apply(lambda x:x.split())

df3['overview'] = df3['overview'].fillna('').astype('str').apply(lambda x:x.split())
df3
```

```python
#Concatinating these columns to give a brief explation/tags of the partticular movie
#Each movie have seperate tags, but we know that similar kind of movies tag line will more closer
# a similar genres movies tagline will be more closer to another same genres movie, the similarity varies for each movie from closer to far

df3['tags'] = df3['overview'] + df3['genres'] + df3['keywords'] + df3['cast'] + df3['director']
```

```python
df3 = df3.drop(columns=['overview','genres','keywords','cast','director'])
```

```python
#the seprate each string having ' ' <space> joining and converting to a lengthy sentences
df3['tags']  = df3['tags'].apply(lambda x: ' '.join(x))          
```

```python
vector = cv.fit_transform(df3['tags'])
```

```python
vector.shape                              # and the 4803 movie datas situated on a  5000 dimensional space
```

```python
#We discussed with the similarities of taglines of each movies,
#We can measure them using this cosine_similarity, it tells the how much closely laid other datas and the 4803 movie datas situated on a  5000 dimensional space

from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Similarity score tells how far is the other movie data points scattered from the current movie data in the scale of-1 to 1
#cos θ =1 , θ=0°, same line/maimum closest point
#cos θ =0 , θ=90°, other data point laid in 90° perpindicular
#cos θ =-1 , θ=180° data laid n directly opposite to the current point

#So we can see that the value is closer to 1 is the nearest point according to the current data

similarity = cosine_similarity(vector)
similarity
```

```python
index = df3[df3['title'] == 'The Lego Movie'].index[0]
# to find the index number for the movie, and its in a tuple using [0] indexing we can fetch the first element from tht index tuple
index
```

```python
#To recommend 10 similsr movies, after getting a movie name

#1. It calculates the most nearest similarity scores and 
#2. giving the index values (normally, it sorted descending with the index)
#3. We need descendingly sorted the movie's index(reversed) based on  similarity scores(enumerate uses for that)
#4. and we fetching [1]th column of each tuple by passing it through key 

#5. In the for loop, fetching only from [1]th data to[11]th data from the distance it will iterate for 10 times
# In every time in the reversly sorted similarity scores corresponded movie index's title will printed

def recommend(movie):
    index = df3[df3['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(df3.iloc[i[0]].title)
        
```

```python
#In the above function enumerates function is used because, when we sorting based on most similar movies it tends to loose the movie's own index number
#enumerate fn. does the job, it retain all movies  similarity scores without changing its index number
#(enumerate fn. converting into list of tuples)

distances = sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])
distances
```

```python
recommend('Titanic')
```

```python
#Its time to make pivot table
#We are making a new data frame having each row of  movie data point represented by the valid users, valid users will be in different columns

df9 = df8.pivot_table(columns= 'userId', index = 'tittle', values = 'rating')
df9.fillna(0,inplace = True)
df9
```

```python
#We can see that for a single movie there will be so many column having 0 rating values, values 0 does not tell anything
#So sparse matrix only contain the compressed rating values which only have values other than 0, so itcan easily implement a movie, using 0 does not give result so sparse matrix simply avoid 0 values

from scipy.sparse import csr_matrix
```

```python
sparx = csr_matrix(df9)
sparx                                 
```
